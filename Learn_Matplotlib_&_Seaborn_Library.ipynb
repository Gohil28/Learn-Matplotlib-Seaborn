```python

# Step 2: Create Subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Step 3: Plot on Subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x
y4 = x**2

ax[0, 0].plot(x, y1, label='sin(x)')
ax[0, 0].set_title('Sin(x)')

ax[0, 1].plot(x, y2, label='cos(x)', color='orange')
ax[0, 1].set_title('Cos(x)')

ax[1, 0].plot(x, y3, label='linear', color='green')
ax[1, 0].set_title('Linear Function')

ax[1, 1].plot(x, y4, label='quadratic', color='red')
ax[1, 1].set_title('Quadratic Function')

# Step 4: Adjust Layout
plt.tight_layout()

# Step 5: Show the Plot
plt.show()
```
